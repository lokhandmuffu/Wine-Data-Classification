```python
from sklearn import svm
import numpy as np
from sklearn.model_selection import train_test_split

X = pd.DataFrame(df, columns=[12, 13])
print(X)
y = df[0]
# x_train, y_train, x_test, y_test = train_test_split(X, y, test_size=0.21, random_state = 0)
print(X.iloc[:,0])
print(X.iloc[:,1])
# fit the model
clf = svm.SVC(kernel='linear', decision_function_shape='ovr', C=1) # C is important here
clf.fit(X, y)
# plt.figure(figsize=(6, 6))
# # get the separating hyperplane
# w = clf.coef_[0]
# a = -w[0] / w[1]
# xx = np.linspace(-1000, 10000)
# yy = a * xx - (clf.intercept_[0]) / w[1]
# plt.figure(1, figsize=(4, 3))
# plt.clf()
# plt.plot(xx, yy, "k-") #********* This is the separator line ************

# plt.scatter(X[12], X[13], c=Y, zorder=10, cmap=plt.cm.Paired,edgecolors="k")
# plt.xlim((0, 4))  
# plt.ylim((0, 1600)) 
# plt.show()

h = 0.02  
# step size in the mesh
# create a mesh to plot in
x_min, x_max = X.iloc[:, 0].min() - 1, X.iloc[:, 0].max() + 1
y_min, y_max = X.iloc[:, 1].min() - 1, X.iloc[:, 1].max() + 1
xx, yy = np.meshgrid(np.arange(x_min, x_max, h),
                     np.arange(y_min, y_max, h))

# title for the plots
titles = ['SVC with linear kernel',
          'LinearSVC (linear kernel)',
          'SVC with RBF kernel',
          'SVC with polynomial (degree 3) kernel']

    # Plot the decision boundary. For that, we will assign a color to each
    # point in the mesh [x_min, x_max]x[y_min, y_max].
plt.subplot(1, 1, 1)
plt.subplots_adjust(wspace=2, hspace=2)

Z = clf.predict(np.c_[xx.ravel(), yy.ravel()])

# Put the result into a color plot
Z = Z.reshape(xx.shape)
plt.contourf(xx, yy, Z, cmap=plt.cm.coolwarm, alpha=0.8)

# Plot also the training points
print()
plt.scatter(X.iloc[:, 0], X.iloc[:, 1], c=y, cmap=plt.cm.coolwarm)
plt.xlabel('OD280/OD315 of diluted wines')
plt.ylabel('Proline')
plt.xlim(xx.min(), xx.max())
plt.ylim(yy.min(), yy.max())
plt.xticks(())
plt.yticks(())
# plt.title(titles[i])

plt.show()
```

```python
I=clf.intercept_
I
```

```python
W=clf.coef_[0]
W

# First Equation coeffients
a = -W[0]/W[1]
b = I[0]/W[1]
```

```python
print(a, "* x + (", b,")")
```

```python
#Second Equation coefficients
W=clf.coef_[1]
W

a = -W[0]/W[1]
b = I[1]/W[1]
```

```python
print(a, "* x + (", b,")")
```

```python
# Third Equation coeffients
W=clf.coef_[2]
W

a = -W[0]/W[1]
b = I[2]/W[1]
```

```python
print(a, "* x + (", b,")")
```

```python
from sklearn.metrics import zero_one_loss

y_pred = clf.predict(X)
accuracy = zero_one_loss(y, y_pred, normalize = False)
# error_rate = 1 - accuracy
print('Number of Misclassifications')
accuracy
```
